```python
df_r = data_set.copy() # Good practice to avoid SettingWithCopy warnings
# Keep Date_utc as a column and extract the month
df_r['month'] = df_r['Date_utc'].dt.month
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 5), sharex=False, sharey=True)
axs = axs.ravel()
df_r = data_set.copy()

# Users distribution by month
df_r['month'] = df_r['Date_utc'].dt.month
df_r.boxplot(column='PV', by='month', ax=axs[0], flierprops={'markersize': 3, 'alpha': 0.3})
df_r.groupby('month')['PV'].median().plot(style='o-', linewidth=0.8, ax=axs[0])
axs[0].set_ylabel('PV')
axs[0].set_title('Power PV distribution by month', fontsize=9)

# Users distribution by week day
df_r['week_day'] = df_r['Date_utc'].dt.day_of_week + 1
df_r.boxplot(column='PV', by='week_day', ax=axs[1], flierprops={'markersize': 3, 'alpha': 0.3})
df_r.groupby('week_day')['PV'].median().plot(style='o-', linewidth=0.8, ax=axs[1])
axs[1].set_ylabel('Power [kW]')
axs[1].set_title('Power PV distribution by week day', fontsize=9)

# Users distribution by the hour of the day
df_r['hour_day'] = df_r['Date_utc'].dt.hour + 1
df_r.boxplot(column='PV', by='hour_day', ax=axs[2], flierprops={'markersize': 3, 'alpha': 0.3})
df_r.groupby('hour_day')['PV'].median().plot(style='o-', linewidth=0.8, ax=axs[2])
axs[2].set_ylabel('Power [kW]')
axs[2].set_title('Power PV distribution by the hour of the day', fontsize=9)
axs[2].set_xticks(range(1, 25, 2))

# Users distribution by week day and hour of the day
mean_day_hour = df_r.groupby(["week_day", "hour_day"])["PV"].mean()
mean_day_hour.plot(ax=axs[3])
axs[3].set(
    title       = "Mean PV power during week",
    xticks      = [i * 24 for i in range(7)],
    xticklabels = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    xlabel      = "Day and hour",
    ylabel      = "Number of users"
)
axs[3].title.set_size(9)

fig.suptitle("Seasonality plots", fontsize=12)
fig.tight_layout()
plt.savefig("seasonality_plot.pdf", dpi=300)
```
